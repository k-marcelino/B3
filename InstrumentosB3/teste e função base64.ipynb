```python
import requests
import pandas as pd
from pandas import json_normalize
import json
from tqdm import tqdm
from time import sleep
```

```python

url = 'https://arquivos.b3.com.br/tabelas/table/TradeInformationConsolidated/2022-11-07/1'
r = requests.get(url)
data = json_normalize(json.loads(r.content))
data
```

```python
page_count = data['pageCount'][0]
page_count
```

```python
# Convertendo em dataframe
columns =  pd.json_normalize(json.loads(r.content), record_path =['columns'])
columns = columns['friendlyName'].tolist()
columns
```

```python
tempo = 300
tempo//60
```

```python
# Convertendo em dataframe
values =  pd.json_normalize(json.loads(r.content), record_path =['values'])
values.columns = columns
values
```

```python
all = []

for pagina in tqdm(range(1, page_count + 1)):
    url = f'https://arquivos.b3.com.br/tabelas/table/TradeInformationConsolidated/2022-11-04/{pagina}'
    r = requests.get(url)
    if r.headers["content-type"] == 'application/json':
        current = json_normalize(json.loads(r.content), record_path =['values'])
        all.append(current)
    else:
        print(f'Erro na página {pagina}, arquivo recebido : {r.headers["content-type"]}')
        continue
    sleep(2)
    
df = pd.concat(all)
df.columns = columns
df
```

```python
pd.concat(all)
```
